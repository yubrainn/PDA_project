```python
import requests as re
import json
```

```python
def getAd(x, y):
  grid = 'y='+str(y)+'&x='+str(x)
  url = 'https://dapi.kakao.com/v2/local/geo/coord2regioncode.json?'+grid
  headers = {"Authorization": "KakaoAK 8c5a706e040158016e817ab111961473"}
  result = json.loads(str(re.get(url, headers=headers).text))
  return result
```

```python
add1 = getAd(126.971451, 37.575794)
type(add1)
```
